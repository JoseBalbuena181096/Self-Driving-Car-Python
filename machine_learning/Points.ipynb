```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#number of points
n_pts=100
np.random.seed(0)
#random samples from a normal distribution, top points,transponced
top_region=np.array([np.random.normal(10,2,n_pts),np.random.normal(12,2,n_pts)]).T
botton_region=np.array([np.random.normal(5,2,n_pts),np.random.normal(6,2,n_pts)]).T
print(top_region)
print(botton_region)
```

```python
#Create graph
#add subplots with 4 in of wigth and 4 in of height
_,ax=plt.subplots(figsize=(4,4))
#axis.scatter(x,y,color)
#[:,0]--> select first column, [:,1]-->select the second column 
ax.scatter(top_region[:,0],top_region[:,1],color="r")
ax.scatter(botton_region[:,0],botton_region[:,1],color="b")
plt.show()
```
