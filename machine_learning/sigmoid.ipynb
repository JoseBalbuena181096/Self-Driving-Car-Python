```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#definition funtions
def draw(x1,x2):
    ln=plt.plot(x1,x2)

def sigmoid(score):
    return 1/(1+np.exp(-score))
```

```python
n_pts=100
np.random.seed(0)
#create a ones array with n_pts elements
bias=np.ones(n_pts)
top_region=np.array([np.random.normal(10,2,n_pts),np.random.normal(12,2,n_pts),bias]).T
bottom_region=np.array([np.random.normal(5,2,n_pts),np.random.normal(5,2,n_pts),bias]).T
# add the top point with the bottom_region in the only array 
all_points=np.vstack((top_region,bottom_region))
#parameter of the model
#w1*x1+w2*x2+b=0
w1=-0.2
w2=-0.35
b=3.5
line_parameters=np.matrix([w1,w2,b]).T
x1=np.array([bottom_region[:,0].min(),top_region[:,0].max()])
x2=-b/w2+(x1*(-w1/w2))

#Product of matrix all_points and line_parameters
linear_combination=all_points*line_parameters
#funtion activation sigmoid
probabilities=sigmoid(linear_combination)
print(probabilities)
#add the plots and shows
_,ax=plt.subplots(figsize=(4,4))
ax.scatter(top_region[:,0],top_region[:,1],color="r")
ax.scatter(bottom_region[:,0],bottom_region[:,1],color="b")
#draw the line 
draw(x1,x2)
plt.show()
```
